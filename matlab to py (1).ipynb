```python

import matplotlib.pyplot as plt
#t=np.arange(-5,10.1,0.2)
#d=np.zeros(len(t))
 
#d[t>=0]=1
t=[]
start=-5
end=10
step=0.2
x=start
while x<=end:
    t.append(round(x,2))
    x+=step

d=[]
for i in t:
    if i>=0:
        d.append(1)
    else:
        d.append(0)

plt.stem(t,d)
plt.title('unit step signal')
plt.xlabel('time')
plt.ylabel('amplitude')
#plt.grid(True)
plt.show()
```

```python
import math
import matplotlib.pyplot as plt

t=[]
start=-5
end=5
step=1
x=start
while x<=end:
    t.append(x)
    x+=step
e=[]
for i in t:
    e.append(math.exp(i))
plt.plot(t,e)
plt.title('exponential signal')
plt.xlabel('time')
plt.ylabel('amplitude')

plt.show()
```

```python
import math
import matplotlib.pyplot as plt
t=[]
start=-2*math.pi
end=2*math.pi
step=0.1
y=start
while y<=end:
    t.append(y)
    y+=step
x=[]
for i in t:
    x.append(math.sin(i))
plt.plot(t,x)
plt.title('sinusoidal function')

plt.show()
```

```python

import matplotlib.pyplot as plt

t=[]
start=-2*math.pi
end=2*math.pi
step=0.1
x=start
while x<=end:
    t.append(x)
    x+=step
x1=[]
for i in t:
    x1.append(math.sin(i))

x2=[]
for j in t:
    x2.append(math.cos(j))
y=[]
for k in range(len(t)):
    y.append(x1[k]+x2[k])
plt.subplot(3, 1, 1)
plt.plot(t, x1)
plt.title('Signal x1 = sin(t)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t, x2)
plt.title('Signal x2 = cos(t)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, y)
plt.title('Added Signal y = x1 + x2')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt

t=[]
start=-2*math.pi
end=2*math.pi
step=0.1
x=start
while x<=end:
    t.append(x)
    x+=step
x1=[]
for i in t:
    x1.append(math.sin(i))

x2=[]
for j in t:
    x2.append(math.cos(j))
y=[]
for k in range(len(t)):
    y.append(x1[k]*x2[k])
plt.subplot(3, 1, 1)
plt.plot(t, x1)
plt.title('Signal x1 = sin(t)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t, x2)
plt.title('Signal x2 = cos(t)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, y)
plt.title('multiplication of Signal y = x1 * x2')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
plt.show()
```
